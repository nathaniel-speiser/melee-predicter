```python
def make_W(x, xi_s, denom ):
    ns = len(xi_s)
    W = np.zeros((ns,ns))
    for i, xi in enumerate(xi_s):
        W[i,i] = np.exp(np.dot((xi-x), (xi-x).T)/-denom) 
    return W
```

```python
def predict(X, Y, test_x, denom):
    W = make_W(test_x, X, denom)
    
    X_m = np.mat(X).T
    Y_m = np.mat(Y).T
    term1 = np.linalg.pinv(np.matmul(X_m.T , np.matmul(W,X_m)))
    
    term2 = np.matmul(X_m.T, np.matmul(W,Y_m))
    theta = np.matmul(term1,term2)
    #print(theta)
    return float(np.dot(test_x,theta)[0][0])
```
